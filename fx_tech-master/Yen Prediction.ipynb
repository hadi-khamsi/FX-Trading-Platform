```python
final_merged_data['D'] = pd.to_datetime(final_merged_data['D'])

# Sort by date just to ensure time series order
final_merged_data.sort_values(by='D', inplace=True)
```

```python
# Create new features: percentage change (for time series)
final_merged_data['Yen_Change'] = final_merged_data['Close Yen'].pct_change()
final_merged_data['MAGS_Change'] = final_merged_data['Mag Close'].pct_change()
final_merged_data['Yield_Change'] = final_merged_data['Close Yield'].pct_change()

# Create lagged features for historical dependency (e.g., lagged values of 1 day)
final_merged_data['Yen_Lag_1'] = final_merged_data['Yen_Change'].shift(1)
final_merged_data['MAGS_Lag_1'] = final_merged_data['MAGS_Change'].shift(1)
final_merged_data['Yield_Lag_1'] = final_merged_data['Yield_Change'].shift(1)
```

```python
final_merged_data.dropna(inplace=True)
```

```python
X = final_merged_data[['MAGS_Lag_1', 'Yield_Lag_1']]  # Feature set (lagged changes in MAGS and Yield)
y = (final_merged_data['Yen_Change'] > 0).astype(int)  # Binary target (1 if Yen follows, 0 otherwise)
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Time series split (5 folds)
tscv = TimeSeriesSplit(n_splits=5)

# Initialize classifier
model = RandomForestClassifier()

# Loop through time-based folds
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = model.predict(X_test)
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
import pickle

with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model saved successfully!")
```
